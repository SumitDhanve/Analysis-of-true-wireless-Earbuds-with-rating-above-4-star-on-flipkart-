# Importing the libraries required for web scraping 

```python
import re
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")

brand=[]
colors=[]
batteries=[]
Price=[]
original_price=[]
rating=[]
Total_reviews=[]
Discount=[]

for i in range(1,12):
    url=f'https://www.flipkart.com/search?q=headphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.headphone_design%255B%255D%3DEarbud&p%5B%5D=facets.rating%255B%255D%3D4%25E2%2598%2585%2B%2526%2Babove&p%5B%5D=facets.headphone_type%255B%255D%3DTrue%2BWireless&page={i}'
    request_header = {'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'}   
    page = requests.get(url,headers=request_header)
    soup= BeautifulSoup(page.text)
    section= soup.find_all('div',class_= 'slAVV4')
    for i in section:
        t= i.find('a',class_='wjcEIp')
        text= t.text
        #print(text)
        #brands
        brands= re.findall('^\w+',text)
        if brands:
            brand.append(brands[0])
        else:
            brand.append(np.nan)
        #colors
        color=i.find('div',class_= 'NqpwHC')
        if color:
            colors.append(color.text)
        else:
            colors.append(np.nan)
        #batteries
        battery = re.findall('(\d+.( Hrs|Hours|HRS|hrs|hours|H|h))',text)
        if battery:
            batteries.append(battery[0])
        else:
            batteries.append(np.nan)
        ## rating
        rate= i.find('div',class_= 'XQDdHH')
        if rate:
            rating.append(rate.text)
        else:
            rating.append(np.nan)
        #Total_reviews
        reviews= i.find('span',class_= 'Wphh3N')
        if reviews:
            Total_reviews.append(reviews.text)
        else:
            Total_reviews.append(np.nan)
        #Price
        price= i.find('div', class_= 'Nx9bqj')
        if price:
            Price.append(price.text)
        else:
            Price.append(np.nan)
        #original_price
        Or_price= i.find('div', class_= 'yRaY8j')
        if Or_price:
            original_price.append(Or_price.text)
        else:
            original_price.append(np.nan)
        #Discount
        discount=i.find('div', class_= 'UkUFwK')
        if discount:
            Discount.append(discount.text)
        else:
            Discount.append(np.nan)

data={'Brands':brand,
      'Colors':colors,
      'Battery':batteries,
      'Price':Price,
      'Original_price':original_price,
      'Rating':rating,
      'Total_reviews':Total_reviews,
      'Discount':Discount
}

df=pd.DataFrame(data)
df
```
